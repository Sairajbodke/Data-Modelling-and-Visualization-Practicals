```python
# Step 4 - Convert the Data into a Unified Format.
```

```python
csv_data.columns = [col.lower() for col in csv_data.columns]
csv_data['date'] = pd.to_datetime(csv_data['date'])
csv_data['unit price'] = csv_data['unit price'].astype(float)
csv_data['quantity'] = csv_data['quantity'].astype(int)
csv_data['total'] = csv_data['total'].astype(float)

excel_data.columns = [col.lower() for col in excel_data.columns]
excel_data['date'] = pd.to_datetime(excel_data['date'])
excel_data['unit price'] = excel_data['unit price'].astype(float)
excel_data['quantity'] = excel_data['quantity'].astype(int)
excel_data['total'] = excel_data['total'].astype(float)

json_data.columns = [col.lower() for col in json_data.columns]
json_data['date'] = pd.to_datetime(json_data['date'])
json_data['unit price'] = json_data['unit_price'].astype(float)
json_data['quantity'] = json_data['quantity'].astype(int)
json_data['total'] = json_data['total'].astype(float)
```

```python
print(csv_data.columns)
print(excel_data.columns)
print(json_data.columns)
```

```python
csv_total_sales = csv_data['total'].sum()
excel_total_sales = excel_data['total'].sum()
json_total_sales = json_data['total'].sum()

```

```python
csv_average_order_value = csv_data['total'].mean()
excel_average_order_value = excel_data['total'].mean()
json_average_order_value = json_data['total'].mean()
```

```python
csv_average_order_value
excel_average_order_value
json_average_order_value
```

```python
csv_category_sales = csv_data.groupby('product line')['total'].sum()
excel_category_sales = excel_data.groupby('product line')['total'].sum()
json_category_sales = json_data.groupby('product_line')['total'].sum()
```

```python
csv_category_sales
excel_category_sales
json_category_sales
```

```python
# Step 6 -Create Visualizations.
```

```python
import matplotlib.pyplot as plt

csv_category_sales.plot(kind='bar', title='Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.show()

excel_category_sales.plot(kind='bar', title='Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.show()

json_category_sales.plot(kind='bar', title='Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.show()
```

```python
category_distribution = csv_data['product line'].value_counts()
category_distribution.plot(kind='pie', title='Product Category Distribution', autopct='%1.1f%%')
plt.ylabel('')
plt.show()

category_distribution = excel_data['product line'].value_counts()
category_distribution.plot(kind='pie', title='Product Category Distribution', autopct='%1.1f%%')
plt.ylabel('')
plt.show()

category_distribution = json_data['product_line'].value_counts()
category_distribution.plot(kind='pie', title='Product Category Distribution', autopct='%1.1f%%')
plt.ylabel('')
plt.show()
```

```python
csv_data.boxplot(column='total', by='product line', grid=False, rot=45)
plt.title('Sales Distribution by Product Category')
plt.suptitle('')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.show()

excel_data.boxplot(column='total', by='product line', grid=False, rot=45)
plt.title('Sales Distribution by Product Category')
plt.suptitle('')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.show()

json_data.boxplot(column='total', by='product_line', grid=False, rot=45)
plt.title('Sales Distribution by Product Category')
plt.suptitle('')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.show()
```
